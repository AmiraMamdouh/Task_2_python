```python
import json
import pandas as pd
from pandas.io.json import json_normalize
#since the data is stored as a multiple json objects;and load function cannot retreive json file 
# with multiple objects and displays "extra data error"; therefore, we wrap all these objects into a list 
#of dictionaries as follows:
my_list= []
with open('usa_gov_click_data_task2.json') as f:
    for json_obj in f: #reading line by line; as each line has a json object
        my_dict = json.loads(json_obj) #putting each line in a dict object
        my_list.append(my_dict) # appending the currently read object in the list of dict objects 
df=pd.DataFrame(my_list)
#df.rename(columns = {'a:web_browser_op_sys', 'r:from_url', 'u: to_url', 'cy: city', 
                   # 'll:longitude_latitude','tz: time_zone',
                   # 't : time_in', 'hc : time_out'}, inplace=True)
df
```

```python
#df[['web_browser','operating_system']]
x=df.a.str.split(pat="[()]", n=0, expand=True)
part_1=x[[0,1]]
part_1
```

```python
part_1.columns=['web_browser','operating_system']
part_1
```

```python
y= df.r.str.split(pat="[// /]", n=3, expand=True)
part_2=y[[2]]
part_2.columns= ['From_URL']
part_2
```

```python
z= df.u.str.split(pat="[// /]", n=3, expand=True)
part_3=z[[2]]
part_3.columns= ['To_URL']
part_3
```

```python
import re
x2= df['ll'].astype(str).str.replace('\[|\]','')
part_4= x2.str.split(pat= ',', n=0, expand= True)
part_4.columns= ['longtude', 'latitude']
part_4
```

```python
import datetime
part_5= pd.to_datetime(df['t'], unit="s")
part_5=part_5.astype(str).str.split(n=0, expand=True)
part_5.columns=['date_in','time_in']
part_5=part_5[['time_in']]
part_5
```

```python
part_6= pd.to_datetime(df['hc'], unit="s")
part_6=part_6.astype(str).str.split(n=0, expand=True)
part_6.columns=['date_out','time_out']
part_6=part_6[['time_out']]
part_6
#part_6.index
```

```python
final_df= pd.concat([part_1,part_2,part_3,df['cy'],part_4,df['tz'],part_5,part_6],axis=1)
final_df.columns=['web_browser','operating_system', 'From_URL', 'To_URL', 'city',
                           'longtude', 'latitude', 'time_zone', 'time_in', 'time_out'] 
final_df
```
